```python
#Import important library
import pandas as pd
import pyodbc
```

```python
#open the data file using pandas
df = pd.read_csv('df1.csv')
df
```

```python
#Now connect python to sql server
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=RON\SQLEXPRESS;'
                      'Database=TestDB;'
                      'Trusted_Connection=yes;')
cursor = conn.cursor()

#Create table
cursor.execute('CREATE TABLE random_test (Date Date(), A int, B int)')

# Insert DataFrame to Table
for row in df.itertuples():
    cursor.execute('''
                INSERT INTO TestDB.dbo.random_test (Date, A, B)
                VALUES (?,?,?)
                ''',
                row.Date, 
                row.A,
                row.B
                )
conn.commit()

#check in sql server
```
